```python
import time
import tkinter as tk
from tkinter import messagebox

class PomodoroTimer:
    def __init__(self, master):
        self.master = master
        self.master.title("Pomodoro Timer")
        self.master.geometry("300x150")

        self.minutes = 25
        self.seconds = 0
        self.is_running = False

        self.label = tk.Label(master, text=self.format_time(), font=("Helvetica", 48), pady=20)
        self.label.pack()

        self.start_button = tk.Button(master, text="Start", command=self.start_timer, padx=10)
        self.start_button.pack(side=tk.LEFT)

        self.reset_button = tk.Button(master, text="Reset", command=self.reset_timer, padx=10)
        self.reset_button.pack(side=tk.RIGHT)

        self.update_display()

    def start_timer(self):
        if not self.is_running:
            self.is_running = True
            self.run_timer()

    def run_timer(self):
        if self.is_running:
            if self.seconds > 0:
                self.seconds -= 1
            elif self.minutes > 0:
                self.minutes -= 1
                self.seconds = 59
            else:
                self.is_running = False
                self.on_timer_completion()

            self.update_display()
            self.master.after(1000, self.run_timer)

    def reset_timer(self):
        self.is_running = False
        self.minutes = 25
        self.seconds = 0
        self.update_display()

    def update_display(self):
        self.label.config(text=self.format_time())

    def format_time(self):
        return f"{self.minutes:02d}:{self.seconds:02d}"

    def on_timer_completion(self):
        self.play_notification_sound()
        messagebox.showinfo("Pomodoro Completed", "Great job! Take a break.")

    def play_notification_sound(self):
        # Add code to play a notification sound (e.g., using a library like pygame)
        pass

if __name__ == "__main__":
    root = tk.Tk()
    timer = PomodoroTimer(root)
    root.mainloop()
```
